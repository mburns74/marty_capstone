## Conat all Decade Tables To One df

```python
cmhf_master_artist_list = pd.concat([cmhf_1960, cmhf_1970,cmhf_1980, cmhf_1990, cmhf_2000, cmhf_2010, cmhf_2020])
cmhf_master_artist_list
```

## Droping the Nan Image column

```python
cmhf_master_artist_list= cmhf_master_artist_list.drop(columns = ['Image'])
cmhf_master_artist_list
```

## Checking df info and data types

```python
cmhf_master_artist_list.info()
```

## Changing Column Names/Reseting Index

```python
cmhf_master_artist_list.columns = ['Inductee','Year_Inducted','Birth_Date','Death_Date','Notes']
cmhf_master_artist_list.reset_index(drop=True)
```

## Extracting Birth Year and Changing From Object to Integer for years between Birth_Year & Year_Inducted

```python
cmhf_master_artist_list['Birth_Year'] = cmhf_master_artist_list['Birth_Date'].str.extract('(\d{4})').astype(int)
cmhf_master_artist_list
```

## Calculating Gap Between Year Born & Year Inducted - This is will be used to determine greatest amount of time between birth and being recognized by the CMHoF

```python
cmhf_master_artist_list['Gap_Between_Birth_Induction'] = (cmhf_master_artist_list.Year_Inducted)-(cmhf_master_artist_list.Birth_Year)
cmhf_master_artist_list
```

## Dropping Birth_Date & Death_Date columns. 
## Not everyone has a death date so this will not be used in calculations. 
## Also points/ranking calculation will be based on Birth_Year As birthdate gets to granular
## Reset index

```python
cmhf_master_artist_list= cmhf_master_artist_list.drop(columns = ['Birth_Date','Death_Date'])
cmhf_master_artist_list.reset_index(drop=True)
```

## Exporting To csv and reordering columns in csv at the same time

cmhf_master_artist_list.to_csv('../data/cmhf_master_artist_list.csv', columns = ['Inductee', 'Year_Inducted', 'Birth_Year','Gap_Between_Birth_Induction','Notes'])
